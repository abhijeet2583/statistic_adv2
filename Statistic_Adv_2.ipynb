1: What is hypothesis testing in statistics?

- Hypothesis testing is a statistical method used to determine whether there is enough evidence in a sample data to draw conclusions about a population. It is a formal procedure for investigating our ideas about the world using statistics. It is most often used by scientists to test specific predictions, called hypotheses, that arise from theories.

Here's a breakdown of the core idea:

Formulate Hypotheses: You start with two competing statements about a population parameter (like a mean, proportion, or standard deviation).

Null Hypothesis (H₀): This is the statement of "no effect" or "no difference." It's the default assumption that nothing interesting is happening.

Alternative Hypothesis (H₁ or Hₐ): This is the statement that contradicts the null hypothesis. It's what you're trying to find evidence for.

Collect Data: You gather a sample of data from the population you're interested in.

Perform a Statistical Test: You choose an appropriate statistical test based on the type of data and the question you're asking. This test calculates a test statistic (like a t-score, z-score, or F-score) that summarizes the relationship between your data and the null hypothesis.

Determine the P-value: The statistical test also gives you a p-value. The p-value is the probability of observing your sample data (or something more extreme) if the null hypothesis were true.

A small p-value suggests that your data is unlikely to have occurred by random chance if the null hypothesis is true.
Make a Decision: You compare the p-value to a predetermined significance level (alpha, often set at 0.05).

If the p-value is less than or equal to alpha, you reject the null hypothesis. This means you have enough evidence to support the alternative hypothesis.

If the p-value is greater than alpha, you fail to reject the null hypothesis.

3.Explain the significance level in hypothesis testing and its role
  in deciding the outcome of a test.

- The significance level, also denoted by the Greek letter alpha
  (α), is a probability threshold that you set before conducting a hypothesis test. It represents the maximum probability of rejecting the null hypothesis when it is actually true (this is also known as a Type I error).

  In simpler terms, alpha is the level of risk you are willing to take of incorrectly concluding that there is a significant effect or difference when there isn't one.

  Common values for alpha are 0.05 (or 5%), 0.01 (or 1%), and 0.10 (or 10%). A smaller alpha value means you require stronger evidence to reject the null hypothesis.


  Role in Deciding the Outcome: The significance level is crucial in deciding the outcome of a hypothesis test because you compare the p-value (obtained from your statistical test) to the significance level.

  If the p-value ≤ α: You have enough evidence to reject the null hypothesis. This means your observed data is statistically significant at the chosen significance level.

  If the p-value > α: You fail to reject the null hypothesis. This means your observed data is not statistically significant at the chosen significance level. You do not have enough evidence to conclude that there is a significant effect or difference.
  
  Essentially, the significance level acts as a benchmark to determine whether the results of your statistical test are considered statistically significant.

6.Write a Python program to generate a binomial distribution with n=10 and
p=0.5, then plot its histogram.

```python
import numpy as np
import matplotlib.pyplot as plt
n = 10  # number of trials
p = 0.5 # probability of success
size = 1000 # number of experiments
binomial_distribution = np.random.binomial(n, p, size)

# Plot the histogram
plt.hist(binomial_distribution, bins=np.arange(0, n+2)-0.5, rwidth=0.8)
plt.xlabel("Number of successes")
plt.ylabel("Frequency")
plt.title("Binomial Distribution (n=10, p=0.5)")
plt.xticks(range(n+2))
plt.grid(axis='y', alpha=0.75)
plt.show()
```

8.Write a Python script to simulate data from a normal distribution and calculate the 95% confidence interval for its mean. Plot the  data using Matplotlib.




```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

# Simulate data from a normal distribution
# Let's assume a population mean (mu) of 50 and standard deviation (sigma) of 2
mu = 50
sigma = 2
sample_size = 1000  # Number of data points to simulate
simulated_data = np.random.normal(mu, sigma, sample_size)

# Calculate the mean and standard error of the sample data
sample_mean = np.mean(simulated_data)
sample_std = np.std(simulated_data, ddof=1) # Use ddof=1 for sample standard deviation
standard_error = sample_std / np.sqrt(sample_size)

# Calculate the 95% confidence interval for the mean
# For a 95% confidence interval, the alpha value is 0.05, so alpha/2 is 0.025
# We need the z-score for a 95% confidence level, which is 1.96 for a two-tailed test
confidence_level = 0.95
alpha = 1 - confidence_level
z_critical = norm.ppf(1 - alpha/2)

margin_of_error = z_critical * standard_error
confidence_interval = (sample_mean - margin_of_error, sample_mean + margin_of_error)

print(f"Sample Mean: {sample_mean}")
print(f"Sample Standard Deviation: {sample_std}")
print(f"Standard Error: {standard_error}")
print(f"Z-critical value for {confidence_level*100}% confidence: {z_critical}")
print(f"Margin of Error: {margin_of_error}")
print(f"{confidence_level*100}% Confidence Interval for the Mean: {confidence_interval}")

# Plot the simulated data as a histogram
plt.hist(simulated_data, bins=30, density=True, alpha=0.6, color='g')

# Plot the probability density function (PDF) of the normal distribution
xmin, xmax = plt.xlim()
x = np.linspace(xmin, xmax, 100)
p = norm.pdf(x, mu, sigma)
plt.plot(x, p, 'k', linewidth=2)

# Add lines for the sample mean and confidence interval
plt.axvline(sample_mean, color='r', linestyle='dashed', linewidth=2, label=f'Sample Mean: {sample_mean:.2f}')
plt.axvline(confidence_interval[0], color='b', linestyle='dashed', linewidth=2, label=f'CI Lower: {confidence_interval[0]:.2f}')
plt.axvline(confidence_interval[1], color='b', linestyle='dashed', linewidth=2, label=f'CI Upper: {confidence_interval[1]:.2f}')


plt.xlabel("Value")
plt.ylabel("Frequency / Density")
plt.title("Simulated Normal Distribution Data with Confidence Interval")
plt.legend()
plt.grid(True)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

sample_data = [49.1, 50.2, 51.0, 48.7, 50.5, 49.8, 50.3, 50.7, 50.2, 49.6,
               50.1, 49.9, 50.8, 50.4, 48.9, 50.6, 50.0, 49.7, 50.2, 49.5,
               50.1, 50.3, 50.4, 50.5, 50.0, 50.7, 49.3, 49.8, 50.2, 50.9,
               50.3, 50.4, 50.0, 49.7, 50.5, 49.9]

def calculate_z_scores(data):
  """
  Calculates the Z-scores for a given dataset.

  Args:
    data: A list or numpy array of numerical data.

  Returns:
    A numpy array of Z-scores.
  """
  mean = np.mean(data)
  std_dev = np.std(data)
  z_scores = (data - mean) / std_dev
  return z_scores

# Calculate Z-scores for the sample data
z_scores = calculate_z_scores(sample_data)

# Visualize the standardized data using a histogram
plt.hist(z_scores, bins=20, edgecolor='black')
plt.xlabel("Z-score")
plt.ylabel("Frequency")
plt.title("Histogram of Z-scores")
plt.grid(True)
plt.show()
```
